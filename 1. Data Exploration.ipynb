## Working on Price Forecasting DataTime

```python
price_forecast_data.head()
```

```python
price_forecast_data.info()
```

```python
price_forecast_data = price_forecast_data.dropna()
```

```python
price_forecast_data.info()
```

```python
# H1 = 1:00
# H2 = 2:00 and so on
# create a dictionary for this
hour_to_time = {}
for i in range(1, 24):
  key = "H" + str(i)
  value = str(i) + ":00"
  hour_to_time[key] = value

hour_to_time["H24"] = "0:00"

print(hour_to_time)
```

```python
datetime_column = []
for index, row in price_forecast_data.iterrows():
    delivery_day = row['Delivery Day']
    hours = row['Hours']
    time = hour_to_time[hours]
    # print(delivery_day + " "+ time)
    datetime_column.append(delivery_day + " "+ time)
```

```python
price_forecast_data["datetime"] = datetime_column
```
