# Imputation Technique for different parameters

**Here, Price and Demand are the dependent variable and need to be forcasted. Hence we are goiong to use only the data sample that has both price and demand as NOT NAN and drop the rest of the data for the training algorithm.**


```python
# only obtain the data samples where both the demand and samples are not NULL
forecast_data_df = copy_data_df.dropna(subset=["Prices\n(EUR/MWh)", "Demand (MW)"])
forecast_data_df.info()
```

```python
nan_count_per_column = forecast_data_df.isna().sum()
print(nan_count_per_column)
```

Interpolation: Temperature, feelslike, dewpoint, humidity, precipitation, preciprob, snow, snowdepth, windgust, windspeed, winddirection, sealevelpressure, cloudcover, visibility, solarradiation, uvindex, severrisk

**For windgust and severrist, it is better to use the mean value as per the distribution of the variable and there are lots of missing values**

**For other parameters, since there are very few missing we are going to use interpolation**

```python
mean_imputing_column = ['windgust', 'severerisk']
interpolate_imputing_column = ['Temperature', 'feelslike', 'dewpoint', 'humidity', 'precipitation', 'precipprob',
       'snow', 'snowdepth', 'windspeed', 'winddirection',
       'sealevelpressure', 'cloudcover', 'visibility', 'solarradiation',
       'uvindex']
```

```python
forecast_data_df.head()
```

```python
forecast_data_df[forecast_data_df["snow.1"] == 1]["snow"]
```

```python
# fill the mean imputing columns with mean value
for column in mean_imputing_column:
  forecast_data_df[column] = forecast_data_df[column].fillna(forecast_data_df[column].mean())
```

```python
# fill the mean imputing columns with mean value
for column in interpolate_imputing_column:
  forecast_data_df[column] = forecast_data_df[column].interpolate()
```

```python
nan_count_per_column = forecast_data_df.isna().sum()
print(nan_count_per_column)
```

**Hence We can see, we have filled all the missing values**
